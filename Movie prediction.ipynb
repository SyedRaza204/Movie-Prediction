```python
import pandas
from pandas import DataFrame as df
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#file reading
data = pandas.read_csv('cost_revenue_clean.csv')
```

```python
#Data describtion
data.describe()
```

```python
#assigning variables to columns
X = df(data, columns = ['production_budget_usd'])
Y = df(data, columns = ['worldwide_gross_usd'])
```

```python
plt.figure(figsize = (10,6)) #figure size
plt.scatter(X,Y, alpha = 0.3) #scatter plotting, alpha value for 3D result

#Title and x,y labeling
plt.title('Movie Cost vs Global Revenue') 
plt.xlabel('Production budget')
plt.ylabel('World wide gross')

#limitation on axis
plt.ylim(0,3000000000)
plt.xlim(0,450000000)

#preview
plt.show()
```

## y = mx + c 
### m = slope coefficient
### c = intercept

```python
regression = LinearRegression() #Create object
regression.fit(X,Y)
```

```python
regression.coef_ #Value of slope coefficient (m), theta_1
```

```python
regression.intercept_ #Intercept value(c)
```

### y = (3.1115)x + (-7236129)

### Result of plotting

```python
plt.figure(figsize = (10,6)) #figure size
plt.scatter(X,Y, alpha = 0.3) #scatter plotting, alpha value for 3D result
plt.plot(X, regression.predict(X), color = 'red', linewidth = 4) #Plotting a line

#Title and x,y labeling
plt.title('Movie Cost vs Global Revenue') 
plt.xlabel('Production budget')
plt.ylabel('World wide gross')

#limitation on axis
plt.ylim(0,3000000000)
plt.xlim(0,450000000)

#preview
plt.show()
```

# # Regression Result
#### How % Result is accurate.

```python
result = regression.score(X,Y)*100
print(result, "%")
```
